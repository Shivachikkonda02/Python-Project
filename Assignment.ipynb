```python
import re

# Function to extract product count from sellerproductcount column
def extract_product_count(text):
    match = re.search(r'over ([\d,]+)', str(text))
    if match:
        return int(match.group(1).replace(',', ''))
    match = re.search(r'(\d+),?(\d*) results', str(text))
    if match:
        return int(match.group(1) + (match.group(2) if match.group(2) else '0'))
    return None

df["parsed_sellerproductcount"] = df["sellerproductcount"].apply(extract_product_count)
```

```python
df['parsed_sellerproductcount']
```

```python
# Function to extract rating percentage and rating count
def extract_seller_ratings(text):
    match = re.search(r'(\d+)% positive.*\((\d+) ratings\)', str(text))
    if match:
        return int(match.group(1)), int(match.group(2))
    return None, None

df[["parsed_rating_percent", "parsed_rating_count"]] = df["sellerratings"].apply(lambda x: pd.Series(extract_seller_ratings(x)))
```

```python
df[["parsed_rating_percent", "parsed_rating_count"]]
```

```python
# Function to extract email and phone number from sellerdetails
def extract_contact_details(text):
    email_match = re.findall(r'[\w\.-]+@[\w\.-]+\.\w+', str(text))
    phone_match = re.findall(r'\+?\d[\d\-\(\) ]{9,}\d', str(text))
    
    email = email_match[0] if email_match else None
    phone = phone_match[0] if phone_match else None
    return email, phone

df[["email", "phone_number"]] = df["sellerdetails"].apply(lambda x: pd.Series(extract_contact_details(x)))
```

```python
df[["email", "phone_number"]]
```

```python
# Function to extract seller country from businessaddress
def extract_country(text):
    if isinstance(text, str):
        if 'CN' in text or 'China' in text:
            return 'China'
        elif 'DE' in text or 'Germany' in text:
            return 'Germany'
        elif 'US' in text or 'United States' in text or 'USA' in text:
            return 'USA'
    return 'Other'

df["seller_country"] = df["businessaddress"].apply(extract_country)

# Filter out Chinese sellers
df = df[df["seller_country"] != "China"]
```

```python
# Convert Hero Product Ratings columns to numeric
df["hero_product_1_ratings"] = pd.to_numeric(df["Hero Product 1 #ratings"], errors='coerce').fillna(0)
df["hero_product_2_ratings"] = pd.to_numeric(df["Hero Product 2 #ratings"], errors='coerce').fillna(0)

```

```python
# Calculate total hero product ratings
df["total_hero_product_ratings"] = df["hero_product_1_ratings"] + df["hero_product_2_ratings"]
```
